```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../../data/2023_to_present.csv')
print(df.head())
```

```python
active_district4_df = df.loc[(df['SHOOTING'] == 1) & (df['DISTRICT'].isin(['B2', 'B3', 'C11', 'E13']))]
active_districts_df = df.loc[df['SHOOTING'] == 1]
other_crime_district4_df = df.loc[(df['SHOOTING'] == 0) & (df['DISTRICT'].isin(['B2', 'B3', 'C11', 'E13']))]
other_crimes_districts_df = df.loc[df['SHOOTING'] == 0]
```

```python
shootings_in_b2 = len(active_district4_df.loc[df['DISTRICT'] == 'B2'])
shootings_in_b3 = len(active_district4_df.loc[df['DISTRICT'] == 'B3'])
shootings_in_c11 = len(active_district4_df.loc[df['DISTRICT'] == 'C11'])
shootings_in_e13 = len(active_district4_df.loc[df['DISTRICT'] == 'E13'])
shootings_in_other_districts = len(active_districts_df) - len(active_district4_df)
other_crimes_districts = len(other_crimes_districts_df) - len(other_crime_district4_df)
```

```python
plot_df = pd.DataFrame({'Districts':['B2', 'B3', 'C11', 'E13', 'OTHER DISTRICTS'], 'val':[shootings_in_b2, shootings_in_b3, shootings_in_c11, shootings_in_e13, shootings_in_other_districts]})

ax = plot_df.plot.bar(x='Districts', y ='val', rot =0)
ax.set_ylabel('Number of Shootings')
plt.title("Districts and Number of Shootings", color = 'red')
```

The bar chart shows the number of shootings in different police districts in Boston, Massachusetts. The chart displays the number of shootings in four specific districts - B2, B3, C11, and E13 - as well as the number of shootings in all other districts combined. Each district is represented by a separate bar, and the height of each bar corresponds to the number of shootings in that district. By analyzing the chart, we can identify which districts have the highest number of shootings and which districts may require more resources to address the problem of gun violence.

```python
active_B2_df = df.loc[(df['SHOOTING'] == 1) & (df['DISTRICT'] == 'B2')]
other_crimes_B2_df = df.loc[(df['SHOOTING'] == 0) & (df['DISTRICT'] == 'B2')]
all_B2_df = df.loc[(df['DISTRICT'] == 'B2')]

active_B3_df = df.loc[(df['SHOOTING'] == 1) & (df['DISTRICT'] == 'B3')]
other_crimes_B3_df = df.loc[(df['SHOOTING'] == 0) & (df['DISTRICT'] == 'B3')]
all_B3_df = df.loc[(df['DISTRICT'] == 'B3')]

active_C11_df = df.loc[(df['SHOOTING'] == 1) & (df['DISTRICT'] == 'C11')]
other_crimes_C11_df = df.loc[(df['SHOOTING'] == 0) & (df['DISTRICT'] == 'C11')]
all_C11_df = df.loc[(df['DISTRICT'] == 'C11')]

active_E13_df = df.loc[(df['SHOOTING'] == 1) & (df['DISTRICT'] == 'E13')]
other_crimes_E13_df = df.loc[(df['SHOOTING'] == 0) & (df['DISTRICT'] == 'E13')]
all_E13_df = df.loc[(df['DISTRICT'] == 'E13')]

gun_crime = [len(active_B2_df), len(active_B3_df), len(active_C11_df), len(active_E13_df), shootings_in_other_districts]
other_crimes = [len(other_crimes_B2_df), len(other_crimes_B3_df), len(other_crimes_C11_df), len(other_crimes_E13_df), other_crimes_districts]
all_crimes = [len(all_B2_df), len(all_B3_df), len(all_C11_df), len(all_E13_df), len(df)]

gun_crime0 = [len(active_B2_df), len(active_B3_df), len(active_C11_df), len(active_E13_df)]
other_crimes0 = [len(other_crimes_B2_df), len(other_crimes_B3_df), len(other_crimes_C11_df), len(other_crimes_E13_df)]
all_crimes0 = [len(all_B2_df), len(all_B3_df), len(all_C11_df), len(all_E13_df)]

index = ['B2', 'B3', 'C11', 'E13', 'OTHER DISTRICTS']
index1 = ['B2', 'B3', 'C11', 'E13', 'ALL DISTRICTS']

index0 = ['B2', 'B3', 'C11', 'E13']
index10 = ['B2', 'B3', 'C11', 'E13']

plot1_df = pd.DataFrame({'Gun Crime': gun_crime,

                   'Other Crimes': other_crimes}, index=index)
plot2_df = pd.DataFrame({'gun crime': gun_crime,

                   'All Crimes': all_crimes}, index=index1)

plot10_df = pd.DataFrame({'Gun Crime': gun_crime0,

                   'Other Crimes': other_crimes0}, index=index0)
plot20_df = pd.DataFrame({'gun crime': gun_crime0,

                   'All Crimes': all_crimes0}, index=index10)

ax1 = plot1_df.plot.bar(rot=0)
ax1.set_ylabel('Number of Crimes')
plt.title("District4 when compared to Other Districts", color = 'red')


ax1 = plot10_df.plot.bar(rot=0)
ax1.set_ylabel('Number of Crimes')
plt.title("District 4 vs Number of Crimes", color = 'red')

ax3 = plot2_df.plot.bar(rot = 0)
ax3.set_ylabel('Number of Crimes')
plt.title("Districts and Number of Crimes", color = 'red')
```

We create several data frames by filtering the original dataframe (df) based on the values in the 'SHOOTING' and 'DISTRICT' columns. We then calculate the number of gun crimes, other crimes, and all crimes for each district and store them in separate lists. Finally, we create four bar charts using these lists and different index labels.

The first chart compares District 4 to other districts, the second chart compares District 4 to all districts, the third chart shows the number of gun crimes and other crimes for each district, and the fourth chart shows the number of gun crimes and all crimes for each district. The charts are plotted using the matplotlib library.

```python

ax2 = plot1_df.plot(y= "Gun Crime", kind = "pie", rot =0, radius = 1.10)


#plt.legend(index, loc='lower left', title='District', prop={'size': 6})
plt.legend(index, loc='lower left', title='District', bbox_to_anchor=(1.0, 0.5))
plt.title("Ratio of Gun Crime", color = 'red')

ax2 = plot1_df.plot(y= "Other Crimes", kind = "pie", rot =0, radius = 1.10)
plt.legend(index, loc='lower left', title='District', bbox_to_anchor=(1.0, 0.5))
plt.title("Ratio of Other Crime", color = 'red')

# ax4 = plot2_df.plot(y="gun crime", kind = "pie", rot =0, radius = 1.10)
# plt.legend(index, loc='lower left', title='District', bbox_to_anchor=(1.0, 0.5))
# plt.title("Ratio of Gun Crime", color = 'red')
ax4 = plot2_df.plot(y="All Crimes", kind = "pie", rot =0, radius = 1.10)
plt.legend(index, loc='lower left', title='District', bbox_to_anchor=(1.0, 0.5))
plt.title("Ratio of Gun Crime", color = 'red')
```

This code creates three pie charts showing the ratio of gun crimes, other crimes, and all crimes in each district. It uses the filtered data from the previous code to calculate the values for each district and plots them using the "plot" function in Pandas. The pie charts are labeled with the district names using a legend and a title.

The plot is titled "Districts vs Gun Crimes over All Crimes" in red font. The plot helps to visualize the proportion of gun crimes in each district relative to all crimes committed in that district.
