##### c) Number of Parks per Feature in Boston City

This code reads in the park dataset and manipulates the 'features' column by splitting it into individual features and counting the number of parks with each feature.

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('../../../data/parks.csv')
df['features'] = df['Feature(s)'].str.split(',')

# Strip leading and trailing spaces from 'features' column
df['features'] = df['features'].apply(lambda x: [f.strip() for f in x])
df = df.explode('features')

# Group by 'features' and get counts
park_counts = df.groupby('features')['Park'].count()
park_counts.plot(kind='bar')

# Set the plot title and axis labels
plt.title('Number of Parks per Feature', fontsize=20)
plt.xlabel('Feature', fontsize=15)
plt.ylabel('Number of Parks', fontsize=15)

# Display the plot
plt.show()
```

From the plot, we can see that the top 5 features with the most parks are Playground, Basketball, Spray Play, Players Bench, and Tennis. This information can be useful for city planners and park management to understand the popular features among park visitors and make informed decisions on park upgrades and improvements.

Additionally, we can see that there are some features with very few parks, such as Boathouse and Ice Skating, which may indicate opportunities for further development in those areas.

Overall, this analysis provides insights into the distribution of park features and can aid in decision-making for park management and planning.

##### e) The number of shootings inside, near, and outside parks in different police districts

This code uses the Google Maps API and various Python libraries such as Pandas, Matplotlib, and Shapely to analyze crime data in different police districts of Boston and determine the number of shootings that occurred inside, near, and outside parks in each district. The code creates a polygon for each park using the Places API, checks whether a shooting occurred inside or within 0.5 miles of the park, and plots a bar chart showing the number of shootings in each category for each district.

```python
import pandas as pd
import matplotlib.pyplot as plt
import googlemaps
from shapely.geometry import Polygon, Point


gmaps = googlemaps.Client(key="AIzaSyCoUEWpSTy1of0j3KiTHEHErTFvHkXN-ds")

# Load the crime data into pandas DataFrames
crime_data = pd.read_csv("../../../data/police_reports/merged_data.csv")


# Define the police districts and their corresponding names
district_names = {"C11": "Dorchester", "B3": "Mattapan", "E5": "Roslindale", "E13": "Jamaica Plain", "B2": "Roxbury"}

# Adding park as a polygon
def apply_park_polygon(x):
  # Use the Places API to get the park's boundary coordinates
  places_result = gmaps.places(x)

  # Extract the boundary coordinates and create a polygon
  coords = places_result['results'][0]['geometry']['viewport']
  return Polygon([(coords['southwest']['lng'], coords['southwest']['lat']), 
                          (coords['northeast']['lng'], coords['southwest']['lat']), 
                          (coords['northeast']['lng'], coords['northeast']['lat']), 
                          (coords['southwest']['lng'], coords['northeast']['lat'])])

data['Polygon']=data["Park"].apply(apply_park_polygon)

# Define a function to determine whether a crime occurred inside or outside a park
def is_inside_park(lat, lon,district):
  parks_in_district=data[data['Location']==district_names[district]]

  # Create a Point object for the latitude and longitude
  point = Point(lon, lat)

  # Loop over each park
  for _, row in parks_in_district.iterrows():
    # Check if the point is inside or outside the park polygon
    if point.within(row.Polygon):
      return True
      
    return False


# Define a function to determine whether a crime occurred within 0.5 miles of a park
def is_near_park(lat, lon,district):
  parks_in_district=data[data['Location']==district_names[district]]

  # Create a Point object for the latitude and longitude
  point = Point(lon, lat)

  # Loop over each park
  for _, row in parks_in_district.iterrows():
    # Check if the distance between the point and park polygon is less than 0.5 miles
    if point.distance(row.Polygon) < 0.5/69:
      return True
      
  return False



# Loop over each police district
for district in district_names:
    # Extract the crime data for this district
    district_crimes = crime_data[crime_data.DISTRICT == district]

    # Filter the crimes to include only shootings
    district_shootings = district_crimes[district_crimes.SHOOTING == 1]

    # Determine whether each shooting occurred inside or outside a park
    district_shootings["inside_park"] = district_shootings.apply(lambda row: is_inside_park(row.Lat, row.Long,row.DISTRICT), axis=1)

    # Determine whether each shooting occurred within 0.5 miles of a park
    district_shootings["near_park"] = district_shootings.apply(lambda row: is_near_park(row.Lat, row.Long,row.DISTRICT), axis=1)

    # Compute the number of shootings inside, outside, and near parks
    num_shootings_inside_parks = district_shootings.inside_park.sum()
    num_shootings_outside_parks = (~district_shootings.inside_park).sum()
    num_shootings_near_parks = district_shootings.near_park.sum()

    # Plot a bar chart showing the number of shootings inside, outside, and near parks
    plt.figure()
    plt.bar(["Inside park", "Near park", "Outside park"], [num_shootings_inside_parks, num_shootings_near_parks, num_shootings_outside_parks])
    plt.title(f"Number of Shootings in {district_names[district]}")
    plt.xlabel("Location")
    plt.ylabel("Number of Shootings")
    plt.show()
```

The given code analyzes crime data from 2015-2023 in several police districts in terms of shootings occurring inside, near, or outside parks. The code uses the Google Maps API to create polygons representing park boundaries and then determines whether a shooting occurred within a park or within 0.5 miles of a park.

The analysis reveals that, for the most part, there are few incidents of shootings occurring inside parks. However, in Roxbury, there were approximately 20 such cases. More concerning is that a high percentage of shootings occurred within 0.5 miles of a park, although this rate varies across the districts.

It is worth noting that the differences in the number of parks and their density in each district could contribute to the variation in the number of incidents occurring within or near parks.
