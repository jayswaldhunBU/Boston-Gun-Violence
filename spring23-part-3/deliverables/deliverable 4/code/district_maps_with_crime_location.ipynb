```python
import pandas as pd
import folium

import geopandas as gpd

crime_data = pd.read_csv("../../../data/merged_data.csv")
crime_data['Lat'].fillna(crime_data['Lat'].mean(), inplace=True)
crime_data['Long'].fillna(crime_data['Long'].mean(), inplace=True)

# Load the districts shapefile
districts = gpd.read_file('../../../data/boston-massachusetts-police-districts.shp')

# Define the name of the district you want to plot
district_name = 'C11'

# Get the geometry of the district
district_geometry = districts[districts['DISTRICT'] == district_name].geometry.iloc[0]

# Filter the crime_data dataframe to keep only crimes in the given district
district_crimes = crime_data[crime_data['DISTRICT'] == district_name]

# Create a Folium map centered on the given district
m = folium.Map(location=[district_geometry.centroid.y, district_geometry.centroid.x], zoom_start=12)

# Add a marker for each crime to the map
for _, row in district_crimes.iterrows():
  folium.Circle(
        location=[row['Lat'], row['Long']],
        color='red',
        fill=True,
        fill_color='red'
    ).add_to(m)

# Display the map
m
```
