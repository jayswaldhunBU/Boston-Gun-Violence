```python
# read the BTDTrafficData.csv file into a pandas dataframe
Traffic_df = pd.read_csv("../../../data/BTDTrafficData.csv")

values_to_drop = [' WR','DO', 'DR', 'BR', 'BRI', 'RO', 'DOR', 'AL', 'X']

# if Traffic_df['Neighborhood Code'].isin(values_to_drop).any():
#     Traffic_df.drop(columns='Neighborhood Code', inplace=True)

rows_to_drop = Traffic_df[Traffic_df['Neighborhood Code'].isin(values_to_drop)].index

Traffic_df.dropna(subset=['Neighborhood Code'], inplace=True)
Traffic_df.drop(index=rows_to_drop, inplace=True)

other_values_to_drop = ['BOSTON PROPER', 'SOUTH BOSTON', 'EAST BOSTON', 'ALLSTON BRIGHTON', 'CHARLESTOWN', 'WEST ROXBURY', 'HYDE PARK']
other_rows_to_drop = Traffic_df[Traffic_df['Neighborhood Code'].isin(other_values_to_drop)].index

Traffic_df.drop(index=other_rows_to_drop, inplace=True)

print(Traffic_df['Neighborhood Code'].unique())

# count the number of streets in each cell of the "Associated Streets" column
street_counts = Traffic_df['Associated Streets'].str.count(',') + 1

# add the "Street Count" column to the dataframe
Traffic_df['Street Count'] = street_counts

Traffic_df
```

```python
print(Traffic_df['Associated Streets'].unique())
```

```python
# extract all unique street names from the "Associated Streets" column
unique_streets = set(Traffic_df['Associated Streets'].str.split(', ', expand=True).stack())

# initialize dictionary to store counts for each street
street_counts = {}

# iterate over each street name and count occurrences in the "Associated Streets" column
for street in unique_streets:
    street_counts[street] = Traffic_df['Associated Streets'].str.contains(street).sum()

# replace all zero values with one
street_counts = {k: 1 if v == 0 else v for k, v in street_counts.items()}

# print the updated street counts
for street, count in street_counts.items():
    print(street, count)
```

```python
unique_streets = set(Traffic_df['Associated Streets'].str.split(', ', expand=True).stack())

street_counts = {}

for street in unique_streets:
    street_counts[street] = Traffic_df['Associated Streets'].str.contains(street).sum()

one_score_streets = [street for street, count in street_counts.items() if count == 1]

print("Streets with score of one:")
for street in one_score_streets:
    print(street)
```

```python
mask = Traffic_df['Study Type'] == 'AUTOMATED TRAFFICE RECORDING'

# Replace 'AUTOMATED TRAFFICE RECORDING' with 'AUTOMATED TRAFFIC RECORDING'
Traffic_df.loc[mask, 'Study Type'] = 'AUTOMATED TRAFFIC RECORDING'

print(Traffic_df['Neighborhood Code'].unique())
print(Traffic_df['Study Type'].unique())
```

```python
mask = (Traffic_df['Study Type'] == 'AUTOMATED TRAFFIC RECORDING') | (Traffic_df['Study Type'] == 'TURNING MOVEMENT COUNT')

# Filter the data frame to keep only the selected rows
Traffic_df = Traffic_df.loc[mask]

study_type_counts = Traffic_df['Study Type'].value_counts()
print(study_type_counts)

fig, ax = plt.subplots(figsize=(12, 6))

wedges, labels = ax.pie(study_type_counts.values, startangle=-40, radius=1)

ax.legend(wedges, ['{} ({:.1f}%)'.format(k, v / sum(study_type_counts.values) * 100) for k, v in study_type_counts.items()], title="Study Type", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
ax.set_title("Percentage of Each Study Type")

plt.savefig('../deliverables/TrafficTypes.png')
plt.show()
```
